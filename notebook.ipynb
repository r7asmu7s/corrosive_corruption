```r
library(ggplot2)
library(data.table)
library(ggthemes)
```

```r
df <- fread('data/hdi_cpi_2020.csv')
colnames(df) <- c('country', 'iso4', 'region', 'cpi_2021', 'cpi_rank_2021', 'sources_2021', 'standard_error_2021', 'hdi_rank_2020', 'hdi_2020')
head(df)
```

```r
pl <- ggplot(df, aes(x = cpi_2021, y = hdi_2020, color = region)) + geom_point(
#     Changing the shape and size of the points in the plot:
    size = 4,
    shape = 1,
) + geom_smooth(
#     Adding the trend line to the plot:
    aes(group = 1),
    method = 'lm',
    formula = y ~ log(x),
    se = FALSE,
    color = 'red',
)

pl
```

The list of the countries to plot was provided by the course tutor.

```r
points_to_label <- c("Russia", "Venezuela", "Iran", "Myanmar", "Sudan",
                   "Afghanistan", "Congo", "Greece", "Argentina", "Brazil",
                   "India", "Italy", "China", "South Africa", "Spain",
                   "Botswana", "Cape Verde", "Bhutan", "Rwanda", "France",
                   "United States", "Turkey", "Germany", "Britain", "Barbados", "Norway", "Japan",
                   "New Zealand", "Singapore")
```

```r
pl <- ggplot(df, aes(x = cpi_2021, y = hdi_2020, color = region)) + geom_point(
#     Changing the shape and size of the points in the plot:
    size = 4,
    shape = 1,
) + geom_smooth(
#     Adding the trend line to the plot:
    aes(group = 1),
    method = 'lm',
    formula = y ~ log(x),
    se = FALSE,
    color = 'red',
) + geom_text(
#     Adding labels to the plot:
    aes(label = country),
    color = 'gray20',
    data = subset(df, country %in% points_to_label),
    check_overlap = TRUE,
)

pl
```

```r
pl <- ggplot(df, aes(x = cpi_2021, y = hdi_2020, color = region)) + geom_point(
#     Changing the shape and size of the points in the plot:
    size = 4,
    shape = 1,
) + geom_smooth(
#     Adding the trend line to the plot:
    aes(group = 1),
    method = 'lm',
    formula = y ~ log(x),
    se = FALSE,
    color = 'red',
) + geom_text(
#     Adding labels to the plot:
    aes(label = country),
    color = 'gray20',
    data = subset(df, country %in% points_to_label),
    check_overlap = TRUE,
) + scale_x_continuous(
#     Changing the x-label:
    name = '2019 Human Development Index (Higher is better)',
) + scale_x_continuous(
#     Changing the x-label label:
    name = '2021 Human Development Index (Higher is better)',
) + ggtitle(
#     Adding title to the plot:
    'Corruption and Human Development',
) + theme_economist()

pl
```
